# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Loading Dataset

```python
df=pd.read_csv('diabetes.csv')
```

# Replace zero value with Mean Value

**A person's Glucose,BloodPressure,SkinThickness,Insulin,Bmi can't be Zero.So,We are replacing it with the respective mean value**

```python
df["Glucose"]=df["Glucose"].replace(0,df["Glucose"].mean())
df["BloodPressure"]=df["BloodPressure"].replace(0,df["BloodPressure"].mean())
df["SkinThickness"]=df["SkinThickness"].replace(0,df["SkinThickness"].mean())
df["Insulin"]=df["Insulin"].replace(0,df["Insulin"].mean())
df["BMI"]=df["BMI"].replace(0,df["BMI"].mean())
```

```python
df.describe()
```

# Distribution Of Outcomes

```python
df['Outcome'].value_counts()
```

# Divide the dataset into X & Y

```python
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

# Apply Feature Scaling

```python
scaler=StandardScaler()
scaler.fit(x)
ss=scaler.transform(x)
```

# Apply Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(ss,y,test_size=0.2,random_state=2)
```

# Training model

# 1.Logistic Regression

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

# Accuracy Score

**Accuracy of Training Data**

```python
x_train_pred=lr.predict(x_train)
train_acc=accuracy_score(x_train_pred,y_train)
print("Accuracy of Training Data : ",train_acc)
```

**Accuracy of Testing Data**

```python
x_test_pred=lr.predict(x_test)
test_acc=accuracy_score(x_test_pred,y_test)
print("Accuracy of Testing Data : ",test_acc)
```

#2. KNeighborsClassifier(KNN)

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

# Accuracy Score

**Accuracy of Training Data**

```python
x_train_pred=knn.predict(x_train)
train_acc=accuracy_score(x_train_pred,y_train)
print("Accuracy of Training Data : ",train_acc)
```

**Accuracy of testing Data**

```python
x_test_pred=knn.predict(x_test)
test_acc=accuracy_score(x_test_pred,y_test)
print("Accuracy of Testing Data : ",test_acc)
```

# 3.Suport Vector Machine

```python
svm=SVC()
svm.fit(x_train,y_train)
```

# Accuracy Score

**Accuracy of Training Data**

```python
x_train_pred=svm.predict(x_train)
train_acc=accuracy_score(x_train_pred,y_train)
print("Accuracy of Training Data : ",train_acc)
```

**Accuracy of Testing data**

```python
x_test_pred=svm.predict(x_test)
test_acc-accuracy_score(x_test_pred,y_test)
print("Accuracy of Testing Data : ",test_acc)
```

# 4.Decision Tree

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

# Accuracy Score

**Accuracy of Training Data**

```python
x_train_pred=dt.predict(x_train)
train_acc=accuracy_score(x_train_pred,y_train)
print("Accuracy of Training Data : ",train_acc)
```

**Accuracy of Testing Data**

```python
x_test_pred=dt.predict(x_test)
test_acc=accuracy_score(x_test_pred,y_test)
print("Accuracy of Testing Data : ",test_acc)
```

# 5.Random Forest

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

# Accuracy Score

**Accuracy of Training Data**

```python
x_train_pred=rf.predict(x_train)
train_acc=accuracy_score(x_train_pred,y_train)
print("Accuracy of Training Data : ",train_acc)
```
